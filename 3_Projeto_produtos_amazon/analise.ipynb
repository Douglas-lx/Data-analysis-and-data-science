### Sobre o conjunto de dados
Este conjunto de dados parece conter informações sobre produtos vendidos na Amazon, incluindo vários atributos, como preços, classificações, disponibilidade e volume de vendas. Aqui está uma descrição de cada coluna.

### Descrições de coluna
**asin:** Número de identificação padrão da Amazon, um identificador exclusivo para cada produto.

**product_title:** O nome ou título do produto.

**product_price:** O preço atual do produto.

**product_original_price:** O preço original do produto antes de quaisquer descontos.

**currency:** A moeda em que o preço do produto é listado.

**product_star_rating:** A classificação média de estrelas do produto.

**product_num_ratings:** O número de classificações que o produto recebeu.

**product_url:** O URL para a página do produto na Amazon.

**product_photo:** Um link para a foto do produto.

**product_num_offers:** O número de ofertas diferentes disponíveis para o produto.

**product_minimum_offer_price:** O preço mínimo das ofertas disponíveis.

**is_best_seller:** Indicador se o produto é um best-seller.

**is_amazon_choice:** Indicador se o produto é um produto Amazon's Choice.

**is_prime:** Indicador se o produto é elegível para Amazon Prime.

**climate_pledge_friendly:** Indicador se o produto é rotulado como Climate Pledge Friendly.

**sales_volume:** O volume de vendas do produto.

**delivery:** Informações sobre as opções de entrega do produto.

**has_variations:** Indicador se o produto tem variações (por exemplo, tamanhos ou cores diferentes).

**product_availability:** O status de disponibilidade do produto.

**unit_price:** O preço por unidade de medida.

**unit_count:** O número de unidades incluídas no preço do produto.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('amazon_product.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(columns= 'Unnamed: 0')
```

```python
df['product_price'] = df.product_price.str.replace('$','')
df['product_price'] = df['product_price'].astype('float')

df['product_minimum_offer_price'] = df.product_minimum_offer_price.str.replace('$','')
df['product_minimum_offer_price'] = df['product_minimum_offer_price'].astype('float')

df['product_original_price'] = df.product_original_price.str.replace('$','')
df['product_original_price'] = df['product_original_price'].astype('float')
```

```python
df.describe()
```

```python
maxPrice = df.loc[:,['product_title','product_price']]
maxPrice.sort_values(by='product_price')
```

```python
qtdChoice = df['is_amazon_choice'].value_counts()
prctChoice = round(df['is_amazon_choice'].value_counts()/len(df['is_amazon_choice']),2)

print(f'{qtdChoice}\n')
print(f'{prctChoice}')
```

```python
fig, ax = plt.subplots(figsize=(6,5))
ax.pie(qtdChoice, labels=['Yes','No'],autopct='%.1f%%',shadow=True,colors=['#47876c','#86c7a9'],explode=(0,0.2))
ax.set_title('Is amazon choice')
plt.savefig('Graficos/amazon_choice.png')
plt.show
```
